```r
# Build a four-factor model
wais.model <- 'verbalcomp =~ vocab + simil + inform + compreh 
workingmemory =~ arith + digspan + lnseq
perceptorg =~ piccomp + block + matrixreason
processing =~ digsym + symbolsearch'

# Analyze the model and include the data argument
wais.fit <- cfa(wais.model, data = IQdata)
```

```r
# Summarize the model with fit.measures and standardized loadings
summary(wais.fit,standardized = TRUE ,fit.measures = TRUE)
```

```r
# Edit the original model
wais.model <- 'verbalcomp =~ vocab + simil + inform + compreh 
workingmemory =~ arith + digspan + lnseq
performance =~ piccomp + block + matrixreason + digsym + symbolsearch'

# Analyze the model and include the model and data argument
wais.fit <- cfa(wais.model, data = IQdata)
```

```r
# Summarize the three-factor model
summary(wais.fit2 , standardized = TRUE, fit.measures = TRUE)
```

```r
# View the fit indices for the updated model
fitmeasures(wais.fit2, c('aic', 'ecvi'))
```

### HIERARCHICAL MODEL

```r
# Update the three-factor model to a hierarchical model
wais.model3 <- 'verbalcomp =~ vocab + simil + inform + compreh 
workingmemory =~ arith + digspan + lnseq
perceptorg =~ piccomp + block + matrixreason + digsym + symbolsearch
simil ~~ inform
general =~ verbalcomp + workingmemory + perceptorg'
# Analyze the hierarchical model where data is IQdata
wais.fit3 <- cfa(wais.model3, data = IQdata)
```

```r
# Examine the fit indices for the old model
fitmeasures(wais.fit2, c('rmsea', 'srmr'))
```

```r
# Examine the fit indices for the new model
fitmeasures(wais.fit3, c('rmsea', 'srmr'))
```

```r
# Update the default picture
semPaths(object = wais.fit3,
         layout = 'tree',
         rotation = 2,
         whatLabels = 'std',
         edge.label.cex = 1,
         what = 'std',
         edge.color = 'navy')
```

A data frame of 75 observations of 11 variables.

y1

    The fairness of elections in 1960

    The effectiveness of the elected legislature in 1960
y5

    Expert ratings of the freedom of the press in 1965
y6

    The freedom of political opposition in 1965
y7

    The fairness of elections in 1965
y8

    The effectiveness of the elected legislature in 1965
x1

    The gross national product (GNP) per capita in 1960
x2

    The inanimate energy consumption per capita in 1960
x3

    The percentage of the labor force in industry in 1960


```r
PolDem.model <- '
  # measurement model
    ind60 =~ x1 + x2 + x3
    dem60 =~ y1 + y2 + y3 + y4
    dem65 =~ y5 + y6 + y7 + y8
  # regressions
    dem60 ~ ind60
    dem65 ~ ind60 + dem60
  # residual correlations
    y1 ~~ y5
    y2 ~~ y4 + y6
    y3 ~~ y7
    y4 ~~ y8
    y6 ~~ y8
'
```

#### Explanations:
- cfa : ind60 is a latent variable and is created from the three manifest variables: x1, x2 and x3
- reg: dem65 is a lin ref function of ind60 and dem60
- y1 is correlated with y5

```r
PolDem.fit <- sem(PolDem.model, data=PoliticalDemocracy)
summary(PolDem.fit, fit.measures = TRUE, standardized=TRUE)
```

```r
semPaths(object = PolDem.fit,
         layout = 'tree',
         rotation = 2,
         whatLabels = 'std',
         edge.label.cex = 1,
         what = 'std',
         edge.color = 'navy')
```

```r
# Examine modification indices 
modificationindices(PolDem.fit, sort = TRUE)
```
